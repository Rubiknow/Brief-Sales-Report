Import packages for the Analysis

```python
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
#Import Sales data 
sales_data = ('/content/work_all_data (1).xlsx')
```

```python
# Convert Excel file to Dataframe using Pandas
df = pd.read_excel(sales_data)
df.head(5)
```

```python
# Lets quicly see the content of our df
print(f'Data contain {df.shape[0]} records and {df.shape[1]} columns.')
```

```python
# Check for Null values
df.isnull().sum()
```

```python
# Drop all null values from the dataset
df1 = df.dropna()
```

```python
df1.isnull().sum()
```

```python
# View Data types
df1.dtypes
```

```python
#Find and delete 'Or'
df1 = df1[df1['Order Date'].str[0:2] != 'Or']
```

```python
# Convert Order Quantity ordered, Price each, order date to the right variables
df1['Quantity Ordered'] = df1['Quantity Ordered'].astype('int')
df1['Price Each'] = df1['Price Each'].astype('float')
df1['Order Date'] = pd.to_datetime(df1['Order Date'])
```

```python
df1.dtypes
```

```python
# Add a month column
df1['Month'] = df1['Order Date'].dt.month
```

```python
df1.head()
```

```python
# Add a month name column
df1['Month Name'] = df1['Order Date'].dt.month_name()
df1.head()
```

```python
# Extract the City and State ID from the purchase address
df1['City'] = df1['Purchase Address'].str.split(",").str.get(1)
df1['State ID'] = df1['Purchase Address'].str.split(" ").str.get(-2)
df1.head()
```

```python
# Extract Hour from the Order Date Column
df1['Hour'] = df1['Order Date'].dt.hour
df1['Day'] = df1['Order Date'].dt.day_name()
df1.head()
```

```python
# Create a Sales Column
df1['Sales'] = df1['Price Each']* df1['Quantity Ordered']
```

```python
df1.describe()
```

# **Data Visualisation**

```python
import plotly.express as px
```

```python
# Top States with Highest Sales

H_sales = df1['Sales'].groupby(df1['City']).sum().to_frame()

fig = px.bar(H_sales, x=H_sales.index, y='Sales', color='Sales', text='Sales')

fig.update_layout(title_text='States With Highest Sales',
                  title_x=0.5, title_font=dict(size=30))


fig.update_traces(textposition='outside', marker=dict(line=dict(color='#000000', width=2)))

```

```python
# Month Sales  
M_sales = df1['Sales'].groupby(df1['Month Name']).sum().sort_values(ascending = False) .to_frame()

fig = px.bar(M_sales, x=M_sales.index, y='Sales', color='Sales', text='Sales')

fig.update_layout(title_text='Total Monthly Sales',
                  title_x=0.5, title_font=dict(size=30))


fig.update_traces(textposition='outside', marker=dict(line=dict(color='#000000', width=2)))
```

```python
# Top Products Sold  
P_sales = df1['Sales'].groupby(df1['Product']).sum().sort_values(ascending = False) .to_frame()

fig = px.bar(P_sales, x=P_sales.index, y='Sales', color='Sales', text='Sales')

fig.update_layout(title_text='Total Products Sold',
                  title_x=0.5, title_font=dict(size=30))


fig.update_traces(textposition='outside', marker=dict(line=dict(color='#000000', width=2)))
```

```python
# Customer Active Time
C_Time = df1['Order ID'].groupby(df1['Hour']).count() .to_frame()

fig = px.line(C_Time, x=C_Time.index, y='Order ID', title='Customer Active Time')
fig.show()



```
